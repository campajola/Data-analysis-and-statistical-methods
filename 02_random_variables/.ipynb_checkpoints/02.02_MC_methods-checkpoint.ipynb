To generate random numbers from a minimum $a$ to a maximum $b$, you can:
- generate a random number with random.random()
- multiply by $b-a$
- add $a$

The result will be in the interval $[a,,b)$.

````python
def rnd_ab(a, b):
    return a+(b-a)*random.random()
for i in range(10):
    print(rnd_ab(-2,2))
````

...the same by giving $a$ and $b$ as arguments of the ```random``` function:

````python
for i in range(10):
    print(random.uniform(-2,2))
````

## Cumulative invertion

````python
import math
````

we generate random number accordinf to a desired pdf by inferting its cumulative

$\displaystyle F(x) = \int_{-\infty}^xf(x^\prime)\,\mathrm{d}x^\prime$

if $r$ in uniform $[0, 1[$ we have to find

$x = F^{-1}(r)$

Take the case of the exponential distribution:
    
$f(x) = \lambda e^{-\lambda x}$


$F(x) = 1 -e^{-\lambda x}$


$\displaystyle x= -\frac{1}{\lambda}\log(1-r)$



````python
def rnd_exp(lambd):
    return -math.log(1-random.random())/lambd
````

````python
N = 100000
lambd = 2
R = [rnd_exp(lambd) for i in range(N)]
````

## convolution

````python
def exp(x):
    return 2*np.exp(-2*x)

sigma = .5
def gaussian(y):
    return 1/math.sqrt(2*math.pi)/sigma*np.exp(-y**2/2/sigma**2)
f = exp
g = gaussian
````

````python
xs = np.linspace(0, 8, 81)
ys = f(xs)
ys/=ys.sum()
````

````python
zs = np.linspace(-4, 4, 81)
ycs = np.convolve(g(zs), ys, 'same')
ycs/=ycs.sum()
````

````python
plt.plot(xs, ys, label='$f(x)$')
plt.plot(xs, ycs, label='$(f\circledast g)(z)$')
plt.xlabel('$x$, $y$')
plt.ylabel('$f(x)$, $(f\circledast g)(y)$')
plt.ylim(0)
plt.legend()
plt.grid()
plt.show()
````

## Legge dei grandi numeri

La legge dei grandi numeri  facile da intuire e ancora pi facile da comprendere con simulazioni al calcolatore. Di seguito la dimostreremo con un esempio.

````python
import matplotlib.pyplot as plt
import random
import numpy as np
import math
````

La legge dei grandi numeri stabilisce una propriet della media di una sequenza di numeri casuali. La media di una sequenza molto grande di numeri casuali "tende" ad un valore fisso, che non  pi casuale, al crescere della dimensione della sequenza.

Simuliamo l'estrazione di un dado con 6 facce con la funzione ```random.randint(1,6)```

````python
r = random.randint(1,6)
print(r)
````

Generiamo un numero alto di lanci

````python
n = 10000
d1 = np.array([random.randint(1,6) for i in range(n)])
d2 = np.array([random.randint(1,6) for i in range(n)])
d3 = np.array([random.randint(1,6) for i in range(n)])
````

````python
#print(list(d1))
````

Definiamo la *media* di $k$ lanci come:

$\bar{x}_k = \frac{\displaystyle 1}{\displaystyle k} {\displaystyle \sum_{i=1}^k x_i}$

Calcoliamo la *media* dei primi $k$ lanci per ogni $k\le n$. A scopo dimostrativo, genereremo due sequenze casuali diverse.

````python
avg1 = [d1[:i+1].mean() for i in range(n)]
avg2 = [d2[:i+1].mean() for i in range(n)]
avg3 = [d3[:i+1].mean() for i in range(n)]
````

Mostriamo in un grafico come cambia $\bar{x}$ al variare di $k$.

````python
plt.figure(figsize=(15,6))
plt.plot(avg1)
plt.plot(avg2)
plt.plot(avg3)
plt.plot([0,n],[3.5,3.5], linestyle='--')
plt.title('Valor medio del lancio simulato di $n$ dadi')
plt.xlabel('n')
plt.ylabel('media')
plt.show()
````

Avevamo gi dimostrato che la varianza di $\bar{x}_k$  pari a $\sigma^2/k$, dove $\sigma^2$  la varianza di tutti gli $x_i$. Ci aspettiamo quindi una *dispersione* della distribuzione di $\bar{x}_k$ sempre minore, man mano che $k$ aumenta.

Aumentando il numero di conteggi $n$  evidente dal grafico che la frequenza, al cerscere di $n$, si avvicina al *valore atteso* di $x$ dalla definizione di probabilit classica:

$\displaystyle \mathbb{E}[x] =\sum_{x=1}^6 x\, P(x) = \sum_{x=1}^6 x\cdot\frac{1}{6} = \frac{1+\cdots +6}{6}=3.5$

````python
np.array(range(1,7)).mean()
````

La legge dei grandi numeri si pu considerare come legge empirica valida per le osservazioni di esperimenti ripetibili e viene usata alla base della formulazione della probabilit intesa in senso *frequentista*.

La formulazione matematica della legge dei grandi numeri pu essere fatta come segue.

$\displaystyle \forall\varepsilon\,\,\,\,\lim_{k\rightarrow\infty}P(|\bar{x}_k-\mathbb{E}[x]|<\varepsilon) = 1$

Ossia, la media del campione di dimensione $k$ *converge in probabilit* al valore atteso $\mathbb{E}[{x}]$. 

Nel caso del lancio dei dadi, questa propriet pu essere provata matematicamente sulla base del calcolo combinatorio, considerando tutti i possibili esiti dei $k$ lanci.

Pi in generale, per, dobbiamo ancora definire il significato di *probabilit* per questi casi, e lo faremo con l'approccio frequentista, che per assume che la legge dei grandi numeri sia valida, con un problema di dipendenza "circolare" del ragionamento.
