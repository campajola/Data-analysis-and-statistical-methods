To generate random numbers from a minimum $a$ to a maximum $b$, you can:
- generate a random number with random.random()
- multiply by $b-a$
- add $a$

The result will be in the interval $[a,,b)$.

````python
def rnd_ab(a, b):
    return a+(b-a)*random.random()
for i in range(10):
    print(rnd_ab(-2,2))
````

...the same by giving $a$ and $b$ as arguments of the ```uniform``` method:

````python
for i in range(10):
    print(random.uniform(-2,2))
````

# Montecarlo methods for random variable generation

# Cumulative invertion

````python
import math
````

we can generate random number according to a desired pdf $f(x)$ by inverting its cumulative:

$\displaystyle F(x) = \int_{-\infty}^xf(x^\prime)\,\mathrm{d}x^\prime$

If $r$ in uniform $[0, 1[$ we have to find

$x = F^{-1}(r)$

````python
# exponential distribution
````

Take the case of the exponential distribution:
    
$f(x) = \lambda e^{-\lambda x}$


$F(x) = 1 -e^{-\lambda x}$


$\displaystyle x= -\frac{1}{\lambda}\log(1-r)$



````python
def rnd_exp(lambd):
    return -math.log(1-random.random())/lambd
````

````python
N = 100000
lambd = 2
R = [rnd_exp(lambd) for i in range(N)]
````

````python
plt.hist(R, bins=100, histtype='step')
plt.xlabel('x')
plt.show()
````

````python
hitst_range=[0, 5]
nbins = 100
plt.hist(R, bins=nbins, range=hitst_range, histtype='step', density=True, label='generated data')

x=np.linspace(hitst_range[0], hitst_range[1], nbins, endpoint=False)
plt.plot(x, lambd*np.exp(-lambd*x), label='lambda*exp(-lambda*x)')

plt.legend()
plt.xlabel('x')
plt.grid()
#plt.yscale('log')
plt.show()
````

````python
hitst_range=[0, 5]
nbins = 100
plt.hist(R, bins=nbins, range=hitst_range, histtype='step', density=True, label='generated data')

x=np.linspace(hitst_range[0], hitst_range[1], nbins, endpoint=False)
plt.plot(x, lambd*np.exp(-lambd*x), label='lambda*exp(-lambda*x)')

plt.legend()
plt.xlabel('x')
plt.grid()
plt.yscale('log')
plt.show()
````

# *hit or miss*

steps:
- we generate random number in two dimensions;
- we do accept only points under the curve we desire.

Do not require cumulative invertion but it is less efficient.

Let's do it with the function:

$f(x)=x\,e^{-x}$

````python
def f_mc(x):
    return x*np.exp(-x)
````

````python
xmin, xmax = 0, 10
x=np.linspace(xmin, xmax, 100)
plt.plot(x,f_mc(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
````

````python
ymax = 0.4
step=0.01
npts=20000
````

````python
xs=np.array([random.random()*(xmax-xmin)+xmin for i in range(npts)])
ys=np.array([random.random()*ymax for i in range(npts)])
````

````python
hits=(ys<f_mc(xs))
````

````python
cols=['red']*npts
for i in range(npts):
    if hits[i]: cols[i]='green'
````

````python
#plt.figure(figsize=(12,12))
x=np.linspace(xmin, xmax, 100)
plt.plot(x,f_mc(x))
plt.scatter(xs,ys,c=cols,s=.1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
````

````python
n_hit = len(hits[hits==True])
f_hit = n_hit/npts
print("n. hit:", n_hit)
print("eff:", f_hit)
````

````python
#plt.figure(figsize=(12,12))
plt.hist(xs[hits==True],histtype='step',density=True,bins=50)
plt.plot(x,f_mc(x),)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
````

## Law of large numbers


````python
import matplotlib.pyplot as plt
import random
import numpy as np
import math
````

The mean of a very large sequence of random numbers "trends" to a fixed value, which is no longer random, as the size of the sequence increases.

Let's simulate the extraction of a die with 6 sides with the function ```random.randint(1,6)```

````python
r = random.randint(1,6)
print(r)
````

let's cosinder n extractions

````python
n = 10000
d1 = np.array([random.randint(1,6) for i in range(n)])
d2 = np.array([random.randint(1,6) for i in range(n)])
d3 = np.array([random.randint(1,6) for i in range(n)])
````

````python
print(list(d1)[:10])
````

We calculate the *average* of the first $k$ tosses for each $k\le n$. For demonstration purposes, we will generate two different random sequences.

````python
avg1 = [d1[:i+1].mean() for i in range(n)]
avg2 = [d2[:i+1].mean() for i in range(n)]
avg3 = [d3[:i+1].mean() for i in range(n)]
````

````python
plt.figure(figsize=(15,6))
plt.plot(avg1)
plt.plot(avg2)
plt.plot(avg3)
plt.plot([0,n],[3.5,3.5], linestyle='--')
plt.title('mean value on n extractions')
plt.xlabel('n')
plt.ylabel('mean value')
plt.show()
````

Increasing the number of counts $n$ it is evident from the graph that the frequency, at the cerscere of $n$, approaches the *expected value* of $x$ from the classic definition of probability:

$\displaystyle \mathbb{E}[x] =\sum_{x=1}^6 x\, P(x) = \sum_{x=1}^6 x\cdot\frac{1}{6} = \frac{1+\cdots +6}{6}=3.5$

## convolution

````python
def exp(x):
    return 2*np.exp(-2*x)

sigma = .5
def gaussian(y):
    return 1/math.sqrt(2*math.pi)/sigma*np.exp(-y**2/2/sigma**2)
f = exp
g = gaussian
````

````python
xs = np.linspace(0, 8, 81)
ys = f(xs)
ys/=ys.sum()
````

````python
zs = np.linspace(-4, 4, 81)
ycs = np.convolve(g(zs), ys, 'same')
ycs/=ycs.sum()
````

````python
plt.plot(xs, ys, label='$f(x)$')
plt.plot(xs, ycs, label='$(f\circledast g)(z)$')
plt.xlabel('$x$, $y$')
plt.ylabel('$f(x)$, $(f\circledast g)(y)$')
plt.ylim(0)
plt.legend()
plt.grid()
plt.show()
````
