# Random Walk


From ```Wikipedia```


In mathematics, a random walk is a random process that describes a path that consists of a succession of random steps on some mathematical space.

An elementary example of a random walk is the random walk on the integer number line Z  which starts at 0, and at each step moves +1 or âˆ’1 with equal probability. 

Other examples include the path traced by a molecule as it travels in a liquid or a gas (see Brownian motion), the search path of a foraging animal, or the price of a fluctuating stock and the financial status of a gambler.

````python
import random
import matplotlib.pyplot as plt
import numpy as np
````

````python
n = 1000 #number of steps

x_0 = 0 #starting point
step=1
x_values = [x_0] #list to store positions
N = [0] #list to store step number

for i in range(n): #'walking' over n steps
    t = random.uniform(0, 1)
    if t>0.5: step_x = step
    else: step_x = -step
    x_values.append(x_values[-1] + step_x)
    N.append(i)

plt.plot(N, x_values, '-')
plt.axhline(y=0, color='red')
plt.scatter(0, x_0, color='red')
plt.title("Random Walk")
plt.xlabel("n", size=14)
plt.ylabel("x", size=14)
plt.grid()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Set up parameters
num_steps = 100  # Number of steps in the random walk
step_size = 1  # Size of each step
start_pos = 0  # Starting position of the random walk

# Create an array to store the positions of the random walk
positions = np.zeros(num_steps)

# Simulate the random walk
for i in range(num_steps):
    positions[i] = start_pos
    start_pos += np.random.choice([-step_size, step_size])

# Set up the plot
fig, ax = plt.subplots()
ax.set_xlim(0, num_steps)
ax.set_ylim(np.min(positions), np.max(positions))
line, = ax.plot([], [])

# Define the animation function
def update(i):
    x = np.arange(i)
    y = positions[:i]
    line.set_data(x, y)
    return line,

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=num_steps, interval=50)

plt.close()
# Display the animation
HTML(ani.to_jshtml())
````

````python
import random
import matplotlib.pyplot as plt

# Definisci la lunghezza del random walk
n = 10000

# Definisci i valori iniziali per la posizione x e y
x_0 = 0
y_0 = 0

# Crea le liste per memorizzare i valori x e y della posizione
x_values = [x_0]
y_values = [y_0]

# Genera il random walk
for i in range(n):
    # Definisci i passi casuali per la posizione x e y
    step_x = random.uniform(-1, 1)
    step_y = random.uniform(-1, 1)

    # Aggiungi i passi alla posizione attuale x e y
    x_values.append(x_values[-1] + step_x)
    y_values.append(y_values[-1] + step_y)

# Crea il grafico del random walk
plt.plot(x_values, y_values)
plt.title("Random Walk in 2D")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````
