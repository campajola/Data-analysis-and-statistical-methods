## generating data

````python
plt.hist(r, bins=25, density=True, histtype='step', label='generated')
x = np.arange(np.min(r), np.max(r), 0.001)
y = expon.pdf(x, scale=tau)
plt.plot(x, y, '--', label='pdf')
plt.vlines(x=r, ymin=0, ymax=0.01, colors='black', lw=1)

plt.annotate(f'events: {size}\nmean: {np.mean(r):.4f}\nstd: {np.std(r):.4f}',
        xy=(0.98, 0.8), xycoords='axes fraction',
        ha='right', va='top', textcoords='offset points', fontsize=12)
    
plt.legend()
plt.grid()
plt.xlabel('x', size=14)
plt.xlabel('f(x)', size=14)
plt.show()
````

# Unbinned maximum likelihhod fit

### Minimum search on grid

````python
# defining likelihood

def neglogl(x,tau):
    return -np.sum(np.log(expon.pdf(x, scale=tau)))
````

````python
steps = 1000

params = np.linspace(tau-3*np.std(r)/np.sqrt(size), tau+3*np.std(r)/np.sqrt(size), steps)

ll = np.array([neglogl(r, p) for p in params])
MLE = params[np.argmin(ll)]
idx_1sigma = np.argwhere(np.diff(np.sign(ll-(neglogl(r,MLE)+1/2)))).flatten()
param_interval = params[idx_1sigma]

dll = (ll[1:]-ll[:-1])/(params[1]-params[0])
d2ll = (dll[1:]-dll[:-1])/(params[1]-params[0])

plt.annotate(f'events: {size}\nmean: {np.mean(r):.3f}\nstd/sqrt(N): {np.std(r)/np.sqrt(size):.3f}\n\
        MLE: {MLE:.3f}\nsqrt(2nd_deriv^-1): {np.sqrt(1/d2ll[np.argmin(ll)]):.3f}\ngraph_error: {param_interval[0]-MLE:.3f}/{param_interval[1]-MLE:.3f}',
        xy=(0.98, 0.98), xycoords='axes fraction',
        ha='right', va='top', textcoords='offset points', fontsize=12)

plt.axvline(x=MLE, linestyle='-', color='red')
plt.axhline(y=neglogl(r,MLE)+1/2, linestyle='--', color='gray')
#plt.axhline(y=neglogl(r,MLE), linestyle='--', color='red')
plt.plot(params, ll)
plt.axvspan(params[idx_1sigma][0], params[idx_1sigma][1], alpha=0.5, color='gray')
#plt.scatter(params[idx], ll[idx], color='gray')
plt.ylim(neglogl(r,MLE)-1/2, neglogl(r,MLE)+2)
plt.xlim(MLE-3*np.std(r)/np.sqrt(size), MLE+3*np.std(r)/np.sqrt(size))
plt.xlabel('tau', size=14)
plt.ylabel('-ln L(x|tau)', size=14)
plt.show()
````

Errors can be estimated by scanning the logl change w.r.t. the parameter. 

1 $\sigma$ (68% errors) corresponds to logL change by 1/2, or $\chi^2$ change by 1 if you use it

````python
plt.hist(r, bins=25, density=True, histtype='step', label='generated')
x = np.arange(np.min(r), np.max(r), 0.001)
y = expon.pdf(x, scale=tau)
plt.plot(x, y, '--', label='pdf')
plt.vlines(x=r, ymin=0, ymax=0.01, colors='black', lw=1)

yfit = expon.pdf(x, scale=MLE)
plt.plot(x, yfit, '--', label='fit')
    
plt.legend()
plt.grid()
plt.xlabel('x', size=14)
plt.xlabel('f(x)', size=14)
plt.show()
````

## minimization with ```scipy```

````python
from scipy.optimize import minimize
````

````python
def fitfunc(tau):
    return neglogl(r, tau)

tau0 = 1.0
res = minimize(fitfunc, x0=tau0)
````

````python
res
````

````python
print("starting value {:.3f}, minimum at {:.3f}".format(tau0, res.x[0]))
````

````python
print("tau = {:.3f} +/- {:.3f}".format(res.x[0], np.sqrt(res.hess_inv[0][0])))
````

## minimization with ```iminuit```

````python
from iminuit import Minuit
````

iminuit is a Python frontend to the Minuit library in C++, an integrated software that combines a local minimizer (called Migrad) and two error calculators (called Hesse and the Minos). You provide it an analytical function, which accepts one or several parameters, and an initial guess of the parameter values. It will then find a local minimum of this function starting from the initial guess. In that regard, iminuit minimizer is like other local minimizers, like those in scipy.optimize.

In addition, iminuit has the ability to compute uncertainty estimates for model parameters. iminuit was designed to solve statistics problems, where uncertainty estimates are an essential part of the result. The two ways of computing uncertainty estimates, Hesse and the Minos, have different advantages and disadvantages.

if you do not use one of the cost functions from the iminuit.cost module, you may need to pass an additional parameter to Minuit.

Minuit by default assumes that the function scales like a chi-square function when one of the parameters is moved away from the minimum. 

If your cost function is constructed as a log-likelihood, it scales differently and you must indicate that to Minuit wit the errordef parameter. Setting this is not needed for the cost functions in iminuit.cost.

The errordef parameter is required to compute correct uncertainties. If you don’t care about uncertainty estimates (but why are you using Minuit then?), you can ignore it. Minuit supports two kinds of cost functions, the negative log-likelihood and the least-squares function. Each has a corresponding value for errordef: - 0.5 or the constant Minuit.LIKELIHOOD for negative log-likelihood functions - 1 or the constant Minuit.LEAST_SQUARES for least-squares functions (the default)

````python
def fitfunc(tau):
    return neglogl(r, tau)
fitfunc.errordef = Minuit.LIKELIHOOD 

tau0 = 1.0
minuit = Minuit(fitfunc, tau=tau0)
minuit.migrad()
````

````python
print("starting value {:.3f}, minimum at {:.3f}".format(tau0, minuit.values["tau"]))
````

````python
for p in minuit.params:
    print(p.name, "= {:.3f} +/- {:.3f}".format(p.value, p.error))
````

Minuit has another algorithm to compute uncertainties: Minos. It implements the so-called profile likelihood method, where the neighborhood around the function minimum is scanned until the contour is found where the function increase by the value of errordef. The contour defines a confidence region that covers the true parameter point with a certain probability. The probability is exactly known in the limit of infinitely large data samples, but approximate for the finite case. Please consult a textbook about statistics about the mathematical details or look at the tutorial “Error computation with HESSE and MINOS”.

````python
#Minos is not automatically called during minimization, it needs to be called explicitly afterwards, like so:

minuit.minos()
````

````python
for p in minuit.params:
    print(p.name, "= {:.3f} {:.3f}/{:.3f}".format(p.value, p.merror[0], p.merror[1]))
````

````python
plt.figure(figsize=(9,6), dpi=100)
x, y = minuit.draw_profile("tau")
````

# a more complicated case

````python
from iminuit import cost
from iminuit import Minuit
import numpy as np
from matplotlib import pyplot as plt
````

````python
from scipy.stats import truncexpon
from scipy.stats import truncnorm
from scipy.stats import norm
````

````python
xMin = 0.
xMax = 1.
````

````python
s = np.random.normal(0.5, 0.1, size=1000)
b = np.random.exponential(1, size=1000)
b = b[b < 1]
x = np.append(s, b)
````

````python
truth = len(s) / len(x), 0.5, 0.1, 1
n, xe, patch = plt.hist(x, bins=25, label='data')
xvalues = np.linspace(np.min(xe), np.max(xe), 100)
#plt.plot(xvalues, model(xvalues, *truth)*len(x)*1/50, '--', color='red', label='model')
plt.legend()
plt.xlabel('x')
plt.ylabel('Entries')
````

## Unbinned

````python
def model(x, f, mu, sigma, slope):
    sig_m = truncnorm.pdf(x, a=(xMin-mu)/sigma, b=(xMax-mu)/sigma, loc=mu, scale=sigma)
    bkg_m = truncexpon.pdf(x, b=(xMax-xMin)/slope, loc=xMin, scale=slope)
    return f * sig_m + (1 - f) * bkg_m
````

````python
truth = len(s) / len(x), 0.5, 0.1, 1
n, xe, patch = plt.hist(x, bins=50, label='data')
xvalues = np.linspace(np.min(xe), np.max(xe), 100)
plt.plot(xvalues, model(xvalues, *truth)*len(x)*1/50, '--', color='red', label='model')
plt.legend()
plt.xlabel('x')
plt.ylabel('Entries')
````

````python
random = 0.5, 0.25, 0.3, 0.51
````

The iminuit package comes with a couple of common cost functions that you can import from iminuit.cost for convenience.

````python
c = cost.UnbinnedNLL(x, model)
m = Minuit(c, *random)
````

````python
m.limits["f", "mu"] = (0, 1)
m.limits["sigma", "slope"] = (0, None)
````

````python
m.interactive(model_points=1000)
````

````python
m.migrad()
````

````python
for p in m.params:
    print(p.name, "= {:.3f} +/- {:.3f}".format(p.value, p.error))
````

````python
print(truth)
````

To get a generic overview, use the method Minuit.draw_mnmatrix. It shows scans over the likelihood where all other parameters than the ones scanned are minimized, in other words, it is using the Minos algorithm. The regions and intervals found in this way correspond to uncertainty intervals. It is also a great way to see whether the likelihood is sane around the minimum.

````python
m.draw_mnmatrix();
````

````python
def model(x, s, mu, sigma, b, slope):
    sig_m = truncnorm.pdf(x, a=(xMin-mu)/sigma, b=(xMax-mu)/sigma, loc=mu, scale=sigma)
    bkg_m = truncexpon.pdf(x, b=(xMax-xMin)/slope, loc=xMin, scale=slope)
    return s + b, s * sig_m + b * bkg_m

c = cost.ExtendedUnbinnedNLL(x, model)
m = Minuit(c, *random)
m.limits["mu"] = (0, 1)
m.limits["sigma", "slope", "s", "b"] = (0, None)

m.interactive()
````
