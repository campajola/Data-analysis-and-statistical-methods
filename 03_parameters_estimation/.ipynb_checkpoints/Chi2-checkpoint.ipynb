````python
import matplotlib.pyplot as plt
import numpy as np
import math
````

### generating data

````python
def func(x, a, b):
    return a + b*x
````

````python
n_points = 20
x_min = 0.0; x_max = 10.0
sigma = 1.8
a0 = 3.0; b0 = 1.5

np.random.seed(1728)
xdata = np.sort(np.random.uniform(x_min, x_max, size=n_points))
ynoise = sigma * np.random.normal(size=xdata.size)
y = func(xdata, a0, b0)
ydata = y + ynoise
sigmas = np.array([sigma]*xdata.size)

plt.errorbar(x=xdata, y=ydata, yerr=sigmas, fmt='o', markersize=4, label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````

## minimization in in closed form

````python
x_mean, y_mean = xdata.mean(), ydata.mean()
dx, dy = xdata - x_mean, ydata - y_mean
dxdy = dx*dy; dx2 = dx*dx
````

Calcolo dei coefficienti del fit

````python
dxdy_mean, dx2_mean = dxdy.mean(), dx2.mean()
b_fit = dxdy_mean/dx2_mean
a_fit = y_mean - b_fit*x_mean
print("a, b = ", a_fit, b_fit)
````

````python
def fit_line(x):
    return a_fit + x*b_fit
````

Calcolo dei valori interpolati e dei residui

````python
y_fit = fit_line(xdata)
res = ydata - y_fit
res2 = res*res
````

Calcoliamo la stima degli errori

````python
x2 = xdata*xdata
x2_mean = x2.mean()
res2_mean = res2.mean()
chi2 = res2_mean*n_points/(sigma*sigma)
ndof = n_points-2
#var_b = res2_mean/dx2_mean/(n_points-2)
var_b = (1/np.sum((1/sigmas)**2))/dx2_mean
res2_mean/dx2_mean/(n_points-2)
var_a = var_b*x2_mean
cov_ab = -var_b*x_mean
err_b = math.sqrt(var_b)
err_a = math.sqrt(var_a)
print('a = {:.3f} ± {:.3f}'.format(a_fit, err_a))
print('b = {:.3f} ± {:.3f}'.format(b_fit, err_b))
print('cov(a,b) = {:.8f}'.format(cov_ab))
print('chi^2 = {:.3f}/{:d} ndof'.format(chi2, ndof))
#print('p = {:.3f}'.format(1 - stats.chi2.cdf(chi2, ndof)))
````

````python
#plt.fill_between(xdata, y_lo_fit, y_hi_fit, facecolor='red', alpha=0.25)
#plt.fill_between(xdata, y_lo3_fit, y_hi3_fit, facecolor='red', alpha=0.1)
plt.errorbar(xdata, ydata, yerr=sigma, fmt='o', label='data')
plt.plot(xdata, func(xdata, a_fit, b_fit), color='red', label='bestfit')

A = 'a = {:.3f} ± {:.3f}'.format(a_fit, err_a)
B = 'b = {:.3f} ± {:.3f}'.format(b_fit, err_b)

plt.annotate(A+'\n'+B,
        xy=(0.02, 0.8), xycoords='axes fraction',
        ha='left', va='top', textcoords='offset points', fontsize=12)

plt.title('analytical regress')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

````python
plt.figure(figsize=(6.5,4), dpi=100)


#plt.figure(figsize=(8,6), dpi=100)

plt.errorbar(xdata, res/sigmas, yerr=1, fmt='o', label='data')
plt.axhline(y=0, c="black", linewidth=2, zorder=0)

plt.title('analytical regress')
plt.grid()
plt.ylim(-5, 5)

plt.xlabel('x')
plt.ylabel('residual/sigma')
plt.show()
````

## minimizzazione con ```scipy```


La libreria ```SciPy``` fornisce la funzione [```curve_fit```](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html) che implementa una minimizzazione numerica. È possibile passare un vettore con gli errori oppure l'intera matrice di covarianza (```sigma```).

Per usare calcolare l'errore dalle derivate seconde del $\chi^2$ bisogna abilitare l'opzione ```absolute_sigma```, che di default è ```False```.</span>


### ```curvefit```

````python
from scipy.optimize import curve_fit
````

````python
#plt.figure(figsize=(9,6), dpi=100)
popt, pcov = curve_fit(func, xdata, ydata, p0=[2, 1], sigma=sigmas, absolute_sigma=True)
print('fit without bounds')
print('--------------------------')
print('curve fit output:')
print('--------------------------')
print('optimal values: \n', popt)
print('covariance matrix: \n', pcov)
print('--------------------------')
print('Parametere estimators')
print("a = {:.3f}+/-{:.3f}".format(popt[0],math.sqrt(pcov[0,0])))
print("b = {:.3f}+/-{:.3f}".format(popt[1],math.sqrt(pcov[1,1])))

a, b = popt[0], popt[1]
var_a, cov_ab, var_b = pcov[0,0], pcov[0,1], pcov[1,1]
y_fit = func(xdata, a, b)

plt.errorbar(xdata, ydata, yerr=sigma, fmt='o', label='data')
plt.plot(xdata, y_fit, color='red', label='bestfit')

A = 'a = {:.3f} ± {:.3f}'.format(popt[0],math.sqrt(pcov[0,0]))
B = 'b = {:.3f} ± {:.3f}'.format(popt[1],math.sqrt(pcov[1,1]))

plt.annotate(A+'\n'+B,
        xy=(0.02, 0.8), xycoords='axes fraction',
        ha='left', va='top', textcoords='offset points', fontsize=12)

plt.title('analytical regress')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

### ```iminuit```

iminuit is a Python frontend to the Minuit library in C++, an integrated software that combines a local minimizer (called Migrad) and two error calculators (called Hesse and the Minos). You provide it an analytical function, which accepts one or several parameters, and an initial guess of the parameter values. It will then find a local minimum of this function starting from the initial guess. In that regard, iminuit minimizer is like other local minimizers, like those in scipy.optimize.

In addition, iminuit has the ability to compute uncertainty estimates for model parameters. iminuit was designed to solve statistics problems, where uncertainty estimates are an essential part of the result. The two ways of computing uncertainty estimates, Hesse and the Minos, have different advantages and disadvantages.

````python
from iminuit import Minuit
````

We use the method of least-squares here, whose cost function computes the sum of squared residuals between the model and the data. The task of iminuit is to find the minimum of that function.

The iminuit module provides a LeastSquares class to conveniently generate a least-squares cost function.

````python
from iminuit.cost import LeastSquares
least_squares = LeastSquares(xdata, ydata, sigmas, func)
m = Minuit(least_squares, a=0, b=0)

#To minimize a function, one has to create an instance of the Minuit class and pass the function and a starting value for each parameter
````

````python
m.migrad()  # ricerca del minimo secondo la discesa del gradiente
````

````python
plt.errorbar(xdata, ydata, sigmas, fmt="o", color='black', label="data")
plt.plot(xdata, func(xdata, *m.values), label="fit", color='red')
fit_info = [ f"$\\chi^2$ / $n_\\mathrm{{dof}}$ = {m.fval:.1f} / {len(xdata) - m.nfit}" ]
for p, v, e in zip(m.parameters, m.values, m.errors):
    fit_info.append(f"{p} = ${v:.3f} \\pm {e:.3f}$")
plt.legend(title="\n".join(fit_info))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

# fit a distribution

````python
from scipy.stats import norm
````

### generating data

````python
def func(x, n, mu, sigma):
    return n/np.sqrt(2*math.pi)/sigma*np.exp(-((x-mu)/sigma)**2 / 2)
````

````python
size = 400
mu = 20
sigma = 4

np.random.seed(17500)
r = norm.rvs(size=size, loc=mu, scale=sigma)
````

````python
#plt.hist(r, density=True, histtype='step', label='generated')
x = np.arange(np.min(r), np.max(r), 0.001)
xrange = [0, 40]
nbins=40
binsize = (xrange[1]-xrange[0])/nbins

H, bin_edges, c = plt.hist(r, range = xrange, bins=nbins, density=False, histtype='step', label='generated')
y = func(x, size*binsize, mu, sigma)
plt.plot(x, y, '--', label='pdf')
plt.vlines(x=r, ymin=0, ymax=5, colors='black', lw=0.1)

plt.annotate(f'events: {size}\nmean: {np.mean(r):.4f}\nstd: {np.std(r):.4f}\nstd_err: {np.std(r)/np.sqrt(size):.4f}',
        xy=(0.98, 0.8), xycoords='axes fraction',
        ha='right', va='top', textcoords='offset points', fontsize=12)

plt.legend()
plt.grid()
plt.xlabel('x', size=14)
plt.xlabel('f(x)', size=14)
plt.show()
````

````python
binsize
````

````python
#plt.hist(r, density=True, histtype='step', label='generated')
x = np.arange(0, 40, 0.001)
y = func(x, size*binsize, mu, sigma)

xdata = bin_edges[:-1] - (bin_edges[0] - bin_edges[1])/2
ydata = H
sigmas = np.sqrt(ydata)

plt.errorbar(x=xdata, y=H, fmt='o', markersize=4, yerr=sigmas, color='black', label=r'$\varepsilon_i=\sqrt{n_i}$')
plt.plot(x, y, '--', color='darkorange', label='pdf')

plt.legend()
plt.grid()
plt.xlabel('x', size=14)
plt.xlabel('f(x)', size=14)
plt.show()
````

````python
def fitfunc(x, mu, sigma):
    return func(x, size*binsize, mu, sigma)
````

````python
plt.errorbar(x=xdata, y=ydata, fmt='o', markersize=4, yerr=sigmas, color='black', label=r'$\varepsilon_i=\sqrt{n_i}$')

popt, pcov = curve_fit(fitfunc, xdata[ydata>0], ydata[ydata>0], p0=[25, 1], sigma=sigmas[ydata>0], absolute_sigma=True)
print(popt)
print(pcov)
print("mu = {:.3f}+/-{:.3f}".format(popt[0],math.sqrt(pcov[0,0])))
print("sigma = {:.3f}+/-{:.3f}".format(popt[1],math.sqrt(pcov[1,1])))

xs =np.linspace(xdata[0], xdata[-1], 2000)
plt.plot(xs, fitfunc(xs, *popt), color='red',
         label=r'fit')

mu_fit = 'mu = {:.3f} ± {:.3f}'.format(popt[0], math.sqrt(pcov[0,0]))
sigma_fit = 'sigma = {:.3f} ± {:.3f}'.format(popt[1], math.sqrt(pcov[1,1]))

plt.annotate(mu_fit+'\n'+sigma_fit,
        xy=(0.02, 0.98), xycoords='axes fraction',
        ha='left', va='top', textcoords='offset points', fontsize=12)


plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````
