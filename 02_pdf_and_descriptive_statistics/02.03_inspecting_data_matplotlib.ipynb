# Histogram

It is a graph showing the number of observations within each given interval.

In Matplotlib, we use the hist() function to create histograms.

The ```hist()``` function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

Let's import some data


````python
from sklearn import datasets
# import some data to play with
iris = datasets.load_iris()
df=pd.DataFrame(iris.data, columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid'])
df['f_class']=iris.target
````

This data sets consists of 3 different types of irises’ (Setosa, Versicolour, and Virginica) petal and sepal length, stored in a 150x4 numpy.ndarray

The rows being the samples and the columns being: Sepal Length, Sepal Width, Petal Length and Petal Width.

matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)

````python
plt.hist(df.sepal_wid, bins=10, range=[2.5, 3.5])
plt.show()
````

````python
plt.hist(df.sepal_wid, density=True)
plt.grid()
plt.show()
````

densitybool, default: False

If True, draw and return a probability density: each bin will display the bin's raw count divided by the total number of counts and the bin width (density = counts / (sum(counts) * np.diff(bins))), so that the area under the histogram integrates to 1 (np.sum(density * np.diff(bins)) == 1).

````python
plt.hist(df.sepal_wid, cumulative=True)
plt.grid()
plt.show()
````

````python
plt.hist(df.sepal_wid, bins=15, alpha=0.5, label='sepal_wid')
plt.legend()
plt.grid()
plt.show()
````

````python
plt.hist(df.sepal_wid, bins=15, histtype='step', lw=2, label='sepal_wid')
plt.legend()
plt.grid()
plt.show()
````

````python
df.head()
````

````python
plt.hist(df.query('f_class==0').sepal_wid, range=[2, 5], bins=15, histtype='step', lw=2, label='class==0')
plt.legend()
plt.grid()
plt.xlabel('sepal_wid')
plt.show()
````

````python
plt.hist([df.query(f'f_class=={i}').sepal_wid for i in [0, 1, 2]], range=[2, 5], bins=15, lw=2, stacked=True, label=['class 0', 'class 1', 'class 2'])
plt.legend()
plt.grid()
plt.xlabel('sepal_wid')
plt.show()
````

````python
plt.hist(df.sepal_wid, range=[2, 5], bins=15, label='all classes')
plt.legend()
plt.grid()
plt.xlabel('sepal_wid')
plt.show()
````

````python
for cl in [0, 1]:
    plt.hist(df.query(f'f_class=={cl}').sepal_wid, range=[2, 5], bins=15, alpha=0.5, label=f'class=={cl}')
plt.legend()
plt.grid()
plt.xlabel('sepal_wid')
plt.show()
````

````python
for feature in df.columns[:2]:
    for cl in df.f_class.unique():
        plt.hist(df.query(f'f_class=={cl}')[feature], range=[df[feature].min(), df[feature].max()], bins=12, alpha=0.5, lw=2, label=f'class=={cl}')
    plt.legend()
    plt.grid()
    plt.xlabel(feature)
    plt.show()
````

````python
for cl in df.f_class.unique():
    plt.scatter(df.query(f'f_class=={cl}')['sepal_len'], df.query(f'f_class=={cl}')['sepal_wid'], label=f'class=={cl}')
plt.legend()
plt.grid()
plt.xlabel('sepal_len')
plt.ylabel('sepal_wid')
plt.show()
````

````python
for feature in df.columns[2:-1]:
    for cl in df.f_class.unique():
        plt.hist(df.query(f'f_class=={cl}')[feature], range=[df[feature].min(), df[feature].max()], bins=12, alpha=0.5, lw=2, label=f'class=={cl}')
    plt.legend()
    plt.grid()
    plt.xlabel(feature)
    plt.show()
````

## Boxplot

A Box Plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. 

In the box plot, a box is created from the first quartile to the third quartile, an horizontal line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

<img style="width: 700px;" src=http://www.lac.inpe.br/~rafael.santos/Docs/CAP394/resources/interpret_boxplots.png>


<img style="width: 500px;" src=https://builtin.com/sites/www.builtin.com/files/styles/ckeditor_optimize/public/inline-images/1_boxplots.jpg>


````python
plt.boxplot([df.query(f'f_class=={i}').sepal_wid for i in [0, 1, 2]])
plt.xlabel('class')
plt.ylabel('sepal_wid')
# show plot
plt.show()
````

Each filled area extends to represent the entire data range, with optional lines at the mean, the median, the minimum, the maximum, and user-specified quantiles.

````python
plt.violinplot([df.query(f'f_class=={i}').sepal_wid for i in [0, 1, 2]])
plt.xlabel('class')
plt.ylabel('sepal_wid')
# show plot
plt.show()
````
