# Uniform

````python
from scipy.stats import uniform
````

````python
# take statistics indicators
````

````python
print("mean:", uniform.mean())
print("variance:", uniform.var())
print("std. dev.:", uniform.std())
print("median:", uniform.median())
````

````python
# take probability
````

````python
uniform.pdf(0.5)
````

````python
step = 0.01
xmin,xmax = 0,10
x = np.arange(xmin, xmax+step, step)
for a, b in ((1, 4), (2, 6), (3, 8)):
    y = uniform.pdf(x, loc=a, scale=b-a)
    plt.fill_between(x, y, alpha=0.2,linewidth=2)
    plt.plot(x, y)
plt.xlim(0, 10)
plt.xlabel('x', size=16)
plt.ylabel('f(x)', size=16)
plt.ylim(bottom=0)
plt.show()
````

````python
# take cumulative
````

````python
a=1
b=3
step = 0.001
xmin,xmax = a-0.1, b+0.1

x = np.arange(xmin, xmax+step, step)
y = uniform.pdf(x, loc=a, scale=b-a)
plt.fill_between(x, y, alpha=0.2,linewidth=2)
plt.plot(x, y, label='pdf')
plt.plot(x, uniform.cdf(x, loc=a, scale=b-a), label='cumulative')
#plt.xlim(0, 10)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x), F(x)')
plt.ylim(bottom=0)
plt.show()
````

````python
# take quantile
````

````python
uniform.ppf(0.01, loc=a, scale=b-a)
````

````python
uniform.ppf(0.5, loc=a, scale=b-a)
````

````python
uniform.ppf(0.99, loc=a, scale=b-a)
````

# normal

````python
from scipy.stats import norm
````

````python
mu = 0
sigma = 1
````

````python
# take statistics indicators
````

````python
print("mean:", norm.mean(loc=mu, scale=sigma))
print("variance:", norm.var(loc=mu, scale=sigma))
print("std. dev.:", norm.std(loc=mu, scale=sigma))
print("median:", norm.median(loc=mu, scale=sigma))
````

````python
# take probability
````

````python
step = 0.01
xmin,xmax = -5,10
x = np.arange(xmin, xmax+step, step)
for mu,sigma in ((0,1), (3, 0.5), (5, 2)):
    y = norm.pdf(x, loc=mu, scale=sigma)
    plt.fill_between(x, y, alpha=0.2,linewidth=2)
    plt.plot(x, y, label=f'mu={mu},sigma={sigma}')
#plt.xlim(0, 10)
plt.ylim(bottom=0)
plt.xlabel('x', size=16)
plt.ylabel('f(x, mu, sigma)', size=16)
plt.legend(fontsize=14)
plt.show()
````

````python
# take cumulative
````

````python
mu=0
sigma=1
step = 0.001
xmin,xmax = -3, 3

x = np.arange(xmin, xmax+step, step)
y = norm.pdf(x, loc=mu, scale=sigma)
plt.fill_between(x, y, alpha=0.2,linewidth=2)
plt.plot(x, y, label='pdf')
plt.plot(x, norm.cdf(x, loc=mu, scale=sigma), label='cumulative')
#plt.xlim(0, 10)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x), F(x)')
plt.ylim(bottom=0)
plt.grid()
plt.show()
````

````python
def p_z_sigma(z):
    return norm.cdf(z)-norm.cdf(-z)
for z in (1, 2, 3, 4, 5):
    print(f"P(in +/- {z} sigma) = {p_z_sigma(z):.6f}")
````

````python
mu=0
sigma=1
step = 0.001
xmin,xmax = -5, 5
x = np.arange(xmin, xmax+step, step)
y = norm.pdf(x, loc=mu, scale=sigma)
    
for z in (1, 2, 3, 4, 5):
    
    xz = np.arange(mu-z, mu+z+step, step)
    yz = norm.pdf(xz, loc=mu, scale=sigma)
    plt.fill_between(xz, yz, alpha=0.2,linewidth=2, label=f'{p_z_sigma(z)*100:.4f} %')
    plt.legend()
    plt.plot(x, y, label='pdf')
    plt.xlabel('x')
    plt.ylabel('f(x), F(x)')
    plt.ylim(bottom=0)
    plt.grid()
    plt.show()
````

````python
# take quantile
````

````python
norm.ppf(0.50)
````

````python
norm.ppf((1-0.68)/2)
````

````python
norm.ppf(0.95)
````

````python
norm.sf(3)
````

# chi2

````python
from scipy.stats import chi2
````

````python
# take statistics indicators
````

````python
dof = 2
````

````python
print("mean:", chi2.mean(dof))
print("variance:", chi2.var(dof))
print("std. dev.:", chi2.std(dof))
print("median:", chi2.median(dof))
````

````python
# take probability
````

````python
step = 0.01
xmin,xmax = -0, 20
x = np.arange(xmin, xmax+step, step)
for dof in [1, 2, 5, 10]:
    y = chi2.pdf(x, dof)
    plt.fill_between(x, y, alpha=0.2,linewidth=2)
    plt.plot(x, y, label=f'dof = {dof}')
#plt.xlim(0, 10)
plt.ylim(bottom=0)
plt.xlabel('x', size=16)
plt.ylabel('f(x, dof)', size=16)
plt.legend(fontsize=14)
plt.ylim(-0.05, 1)
plt.show()
````

````python
step = 0.01
xmin,xmax = 50, 150
x = np.arange(xmin, xmax+step, step)
for dof in [100]:
    y = chi2.pdf(x, dof)
    plt.fill_between(x, y, alpha=0.2,linewidth=2)
    plt.plot(x, y, label=f'dof = {dof}')
#plt.xlim(0, 10)
plt.ylim(bottom=0)
plt.xlabel('x', size=16)
plt.ylabel('f(x, dof)', size=16)
plt.legend(fontsize=14)
plt.show()
````

````python
# take cumulative
````

````python
dof = 3
step = 0.001
xmin,xmax = -0, 10

x = np.arange(xmin, xmax+step, step)
y = chi2.pdf(x, dof)
plt.fill_between(x, y, alpha=0.2,linewidth=2)
plt.plot(x, y, label='pdf')
plt.plot(x, chi2.cdf(x, dof), label='cumulative')
#plt.xlim(0, 10)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x), F(x)')
plt.ylim(bottom=0)
plt.grid()
plt.show()
````

````python
# take quantile
````

````python
chi2.ppf(0.4, dof)
````

````python
chi2.sf(3, dof)
````

````python
mean, var, skew, kurt = chi2.stats(dof, moments='mvsk')
````
