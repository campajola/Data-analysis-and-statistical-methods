# Law of large numbers


````python
import matplotlib.pyplot as plt
import random
import numpy as np
import math
````

The mean of a very large sequence of random numbers "trends" to a fixed value, which is no longer random, as the size of the sequence increases.

Let's simulate the extraction of a die with 6 sides with the function ```random.randint(1,6)```

````python
r = random.randint(1,6)
print(r)
````

let's cosinder n extractions

````python
n = 1000
d1 = np.array([random.randint(1,6) for i in range(n)])
d2 = np.array([random.randint(1,6) for i in range(n)])
d3 = np.array([random.randint(1,6) for i in range(n)])
````

````python
print(list(d1)[:10])
````

We calculate the *average* of the first $k$ tosses for each $k\le n$. For demonstration purposes, we will generate two different random sequences.

````python
avg1 = [d1[:i+1].mean() for i in range(n)]
avg2 = [d2[:i+1].mean() for i in range(n)]
avg3 = [d3[:i+1].mean() for i in range(n)]
````

````python
plt.figure(figsize=(14,6))
plt.plot(avg1)
plt.plot(avg2)
plt.plot(avg3)
plt.plot([0,n],[3.5,3.5], linestyle='--')
plt.title('mean value on n extractions')
plt.xlabel('n', size=14)
plt.ylabel('mean value', size=14)
plt.show()
````

Increasing the number of counts $n$ it is evident from the graph that the frequency, at the cerscere of $n$, approaches the *expected value* of $x$ from the classic definition of probability:

$\displaystyle \mathbb{E}[x] =\sum_{x=1}^6 x\, P(x) = \sum_{x=1}^6 x\cdot\frac{1}{6} = \frac{1+\cdots +6}{6}=3.5$

# variance of the mean distribution

````python
N=100
#x = np.random.normal(loc=0, scale=1, size=N)
x = np.random.random(size=N)
plt.hist(x, bins=20, histtype='step', label='x')
plt.grid()
plt.xlabel('x')
plt.annotate(f'n entries = {N}', xy=(.05,.9), xycoords='axes fraction', size=12)
plt.annotate(f'mean = {np.mean(x):.3f}', xy=(.05,.85), xycoords='axes fraction', size=12)
plt.annotate(f'std. dev. = {np.std(x):.3f}', xy=(.05,.8), xycoords='axes fraction', size=12)
plt.show()
````

````python
# let's compute the distribution of the averages computed over 100 i.i.d r.v.
````

````python
AVG=[]
for i in range(10000):
    N=100
    x = np.random.normal(loc=0, scale=1, size=N)
    x = np.random.random(size=N)
    AVG.append(np.mean(x))

plt.hist(AVG, range=[-1, 1], bins=100, histtype='step', label='x')
plt.grid()
plt.xlabel('sample mean value')

plt.annotate(f'mean = {np.mean(AVG):.3f}', xy=(.05,.9), xycoords='axes fraction', size=12)
plt.annotate(f'std. dev. = {np.std(AVG):.3f}', xy=(.05,.80), xycoords='axes fraction', size=12)

plt.show()
````

````python
STD_AVG=[]
npop=[1, 10, 100, 200, 500, 1000, 10000]
for n in npop:
    AVG=[]
    for i in range(10000):
        N=n
        x = np.random.normal(loc=0, scale=1, size=N)
        x = np.random.random(size=N)
        AVG.append(np.mean(x))
    plt.hist(AVG, range=[-1, 1], bins=100, density=True, histtype='step', label=f'n={n}')
    STD_AVG+=[np.std(AVG)]
    print(f'average of {n} rv: mean = {np.mean(AVG):.3f}, std. dev. = {np.std(AVG):.3f}')

plt.legend()
plt.grid()
plt.xlabel('mean(x)')
plt.show()
````

````python
plt.scatter(npop, STD_AVG)
x = np.arange(0, np.max(npop), 1)
plt.plot(x, 1/np.sqrt(12)/np.sqrt(x), '--', color='orange', label='1/sqrt(n)')
plt.xlabel('sample size n', size=14)
plt.ylabel('std. dev. of <x>', size=14)
plt.legend()
plt.show()
````

Let X_1 , … , X_n  be a sequence of random samples — that is, a sequence of i.i.d. random variables drawn from a distribution of expected value given by μ  and finite variance given by $σ^2$. 
Suppose we are interested in the sample average


\<X> ≡ $X_1$ + ⋯ + $X_n$ / n

of the first n samples.

By the law of large numbers, the sample averages converge  probability to the expected value μ.

The  central limit theorem describes the size and the distributional form of the stochastic fluctuations around the deterministic number μ  during this convergence. 

For large enough n, the distribution of \<X>  gets arbitrarily close to the normal distribution with mean μ  and variance $σ^2/n$
