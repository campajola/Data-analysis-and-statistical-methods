```python
import random
import matplotlib.pyplot as plt

# Definisci la lunghezza del random walk
n = 10000

# Definisci i valori iniziali per la posizione x e y
x_0 = 0
y_0 = 0

# Crea le liste per memorizzare i valori x e y della posizione
x_values = [x_0]
y_values = [y_0]

# Genera il random walk
for i in range(n):
    # Definisci i passi casuali per la posizione x e y
    step_x = random.uniform(-1, 1)
    step_y = random.uniform(-1, 1)

    # Aggiungi i passi alla posizione attuale x e y
    x_values.append(x_values[-1] + step_x)
    y_values.append(y_values[-1] + step_y)

# Crea il grafico del random walk
plt.plot(x_values, y_values)
plt.title("Random Walk")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Definisci la lunghezza del random walk
n = 10000

# Definisci i valori iniziali per la posizione x e y
x_0 = 0
y_0 = 0

# Crea le liste per memorizzare i valori x e y della posizione
x_values = [x_0]
y_values = [y_0]
step = 1

# Genera il random walk
for i in range(n):
    # Definisci i passi casuali per la posizione x e y
    t = random.uniform(0, 1)*2*np.pi
    step_x = step*np.cos(t)
    step_y = step*np.sin(t)
    # Aggiungi i passi alla posizione attuale x e y
    x_values.append(x_values[-1] + step_x)
    y_values.append(y_values[-1] + step_y)

# Crea il grafico del random walk
plt.plot(x_values, y_values)
plt.title("Random Walk")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Definisci la lunghezza del random walk
n = 1000

# Definisci i valori iniziali per la posizione x e y
x_0 = 0

# Crea le liste per memorizzare i valori x e y della posizione
x_values = [x_0]
step = 1
N = [0]

# Genera il random walk
for i in range(n):
    # Definisci i passi casuali per la posizione x e y
    t = random.uniform(0, 1)
    if t>0.5: step_x = step
    else: step_x = -step
    # Aggiungi i passi alla posizione attuale x e y
    x_values.append(x_values[-1] + step_x)
    N.append(i)

# Crea il grafico del random walk
plt.plot(N, x_values, '-')
plt.title("Random Walk")
plt.xlabel("n")
plt.ylabel("x")
plt.grid()
plt.show()
```
